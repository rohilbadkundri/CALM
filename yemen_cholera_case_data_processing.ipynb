```python
# need to convert data into stationary (changes in cholera cases from one day to another)

# the function below creates a differenced series
def difference(dataset, interval=1):
    
    diff = pd.Series()
    
    for i in range(interval, len(dataset)):
        
        print ('new total cases: ' + str(dataset[i]) + ', old total cases: ' + str(dataset[i-1]))
        print ('new date: ' + str(dataset.index[i]) + ', old date: ' + str(dataset.index[i-1]))
        
        # get the number of new cases in the ith time step
        case_delta = dataset[i] - dataset[i - interval]
        
        # number of days between the ith time step and i-1th time step
        time_delta = (dataset.index[i] - dataset.index[i - interval]).days

        idx = pd.date_range(dataset.index[i - interval], dataset.index[i])[1:]
        
        # normalize the number of new cases against the number of days in between the two time steps
        value = case_delta / time_delta
        
        # linearly interpolating - if there are multiple days between each time cholera is reported
        # we assume that the rate of change is uniform across that time period
        # therefore we repeat the same value for each day
        repeat_value = np.repeat(value, time_delta)
        
        #print (repeat_value, idx)
        print (pd.Series(repeat_value, idx))
        
        diff = diff.append(pd.Series(repeat_value, idx))

        
    return ((diff))
```

```python
def get_cholera_cases_x_days_ahead(row, days, data):
    
    new_cases = data[(data.gov_iso == row.gov_iso) & (data.date > row.date) & (data.date <= (row.date + pd.to_timedelta(days, unit='d')))].new_cases.sum()
    
    return (new_cases)
    
```

```python
def get_cholera_cases_x_days_ahead_week_only(row, days_forward, days_back, data):
    
    new_cases = data[(data.gov_iso == row.gov_iso) & (data.date > (row.date + pd.to_timedelta(days_forward - days_back, unit='d'))) & (data.date <= (row.date + pd.to_timedelta(days_forward, unit='d')))].new_cases.sum()
    
    return (new_cases)
    
```
